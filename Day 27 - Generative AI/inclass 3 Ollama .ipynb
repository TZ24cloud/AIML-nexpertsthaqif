### Prompt engineering with Ollama

```python
import ollama

messages = [
    {"role": "system", "content": "You are a professional editor. Convert slang into formal English."},
    {"role": "user", "content": "That movie was fire!"},        # Example 1
    {"role": "assistant", "content": "That cinema was excellent."}, # Answer 1
    {"role": "user", "content": "I'm ghosting that meeting."},  # The real task
]

# OPTIONS: These are your "knobs" for prompt engineering
response_stream = ollama.chat(
    model='phi3', 
    messages=messages,
    stream=True,
    options={
        'temperature': 0.1,    # Low = focused and predictable; High = creative
        'num_ctx': 4096,       # How much "memory" (context window) to use
        'top_p': 0.9,          # Nucleus sampling; helps with text diversity
        # 'stop': ['User:', '\n'] # Hard stop if the model tries to keep talking
    }
)

# PRINTING: Handling the stream generator
print("Result: ", end="")
for chunk in response_stream:
    print(chunk['message']['content'], end='', flush=True)
```
